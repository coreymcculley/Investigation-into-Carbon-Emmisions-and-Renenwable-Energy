```python
df.drop(columns = {'Project','Investor','Source'}, inplace=True)
```

```python
Countries = ["Austria","Belgium","Bulgaria","Croatia","Cyprus","Czech Republic","Denmark","Estonia","Finland","France","Germany","Greece","Hungary","Ireland",
      "Italy","Latvia","Lithuania","Luxembourg","Malta","Netherlands","Poland","Portugal","Romania","Slovakia","Slovenia","Spain","Sweden","USA","UK","China"]
```

```python
Countries_df = df[df['Recipient Country/Area'].isin(Countries)]
```

```python
Countries_df
```

```python
sum_countries = Countries_df.groupby(['Recipient Country/Area','Year']).sum()
```

```python
sum_countries.head(20)
```

```python
#Use GroupBy in order to aggregate the data according to the values in the country / year
df_countries = df.groupby(['Recipient Country/Area','Year']).sum()
df_countries.head(50)
```
