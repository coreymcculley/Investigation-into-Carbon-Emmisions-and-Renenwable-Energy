#### Read the csv

```python
df_SE4ALL = pd.read_csv(file)
df_SE4ALL.head(10)
```

#### Filter Data to 2015


```python
df_SE4ALL_2015 = df_SE4ALL[df_SE4ALL["Time"] == 2015]
df_SE4ALL_2015.head()
```

#### What 10 developed countries had the highest % use of renewable energy of total final energy consumption (TFEC) in 2015? 



```python
list_of_developed_countries_eu = ["Austria", "Belgium", "Denmark", "Finland", "France", "Germany", "Greece", "Ireland", "Italy", "Luxembourg", "Netherlands", "Portugal", "Spain", "Sweden", "United Kingdom", 
                                  "Bulgaria", "Croatia", "Cyprus", "Czech Republic", "Estonia", "Hungary", "Latvia", "Lithuania", "Malta", "Poland", "Romania", "Slovak Republic", "Slovenia", 
                                  "Iceland", "Norway", "Switzerland"]
list_of_developed_countries_other = ["Australia", "Canada", "Japan", "New Zealand", "United States"]
list_of_developed_countries = list_of_developed_countries_eu + list_of_developed_countries_other

#list of countries with developed economies based on 2014 WESP Country Classification per UN https://www.un.org/en/development/desa/policy/wesp/wesp_current/2014wesp_country_classification.pdf

df_SE4ALL_2015_developed = df_SE4ALL_2015[df_SE4ALL_2015["Country Name"].isin(list_of_developed_countries)]
df_SE4ALL_2015_developed['Renewable energy share of TFEC (%) [2.1_SHARE.TOTAL.RE.IN.TFEC]'] = df_SE4ALL_2015_developed['Renewable energy share of TFEC (%) [2.1_SHARE.TOTAL.RE.IN.TFEC]'].astype(float)

df_SE4ALL_2015_developed.sort_values(by=['Renewable energy share of TFEC (%) [2.1_SHARE.TOTAL.RE.IN.TFEC]'], inplace=True, ascending=False)
```

```python
df_SE4ALL_2015_developed[['Country Name', 'Renewable energy share of TFEC (%) [2.1_SHARE.TOTAL.RE.IN.TFEC]']].head(10)
```

```python
file = os.path.join("Source Databases","GDP_POP_Energy.csv")
df_SE4ALL_GDP = pd.read_csv(file)

#replace all '..' str vales with 0
df_SE4ALL_GDP = df_SE4ALL_GDP.replace('..',0)

list_of_columns = df_SE4ALL_GDP.columns.tolist()

for i in range(8, 18):
    column_name = list_of_columns[i]
    df_SE4ALL_GDP[f'{column_name}'] = df_SE4ALL_GDP[f'{column_name}'].astype(float)
    
    


```

```python
df_SE4ALL_GDP.dtypes
```

```python
df_SE4ALL_GDP["Access to Clean Fuels and Technologies for cooking (% of total population) [2.1_ACCESS.CFT.TOT]"] = df_SE4ALL_GDP["Access to Clean Fuels and Technologies for cooking (% of total population) [2.1_ACCESS.CFT.TOT]"].astype(float)
```

```python
df_SE4ALL_GDP.dtypes
```

```python
df_SE4ALL_GDP["Date"] = df_SE4ALL_GDP["Date"].astype(int)
```

```python
df_SE4ALL_GDP.dtypes
```

```python
#find sum annually of EU RE consumption and TFEC (including UK) from 1990-2015

#limit df to only EU
df_SE4ALL_GDP_eu = df_SE4ALL_GDP[df_SE4ALL_GDP["Country Name"].isin(list_of_developed_countries_eu)]

#limit df to range of years from 1990 to 2015
df_SE4ALL_GDP_eu_years = df_SE4ALL_GDP_eu[df_SE4ALL_GDP_eu["Date"].between(1990, 2015)]

#group by year and sum
df_SE4ALL_GDP_eu_years.groupby("Date").sum()

re_consumption_TJ_eu = df_SE4ALL_GDP_eu_years.groupby("Date").sum()["Renewable energy consumption (TJ) [3.1_RE.CONSUMPTION]"]
tfec_TJ_eu = df_SE4ALL_GDP_eu_years.groupby("Date").sum()["Total final energy consumption (TFEC) (TJ) [1.1_TOTAL.FINAL.ENERGY.CONSUM]"]
gdp_eu = df_SE4ALL_GDP_eu_years.groupby("Date").sum()["GDP"]

re_share_of_TFEC_eu = re_consumption_TJ_eu / tfec_TJ_eu * 100
re_share_of_TFEC_eu

```

```python
#find annual er_consumption_TJ_us, tfec_TJ_us, re_share_of_TFEC_us
#limit df to only US
df_SE4ALL_GDP_us = df_SE4ALL_GDP[df_SE4ALL_GDP["Country Name"] == "United States"]

#limit df to range of years from 1990 to 2015
df_SE4ALL_GDP_us_years = df_SE4ALL_GDP_us[df_SE4ALL_GDP_us["Date"].between(1990, 2015)]

#group by year and grab relevant data
re_consumption_TJ_us = df_SE4ALL_GDP_us_years.groupby("Date").sum()["Renewable energy consumption (TJ) [3.1_RE.CONSUMPTION]"]
tfec_TJ_us = df_SE4ALL_GDP_us_years.groupby("Date").sum()["Total final energy consumption (TFEC) (TJ) [1.1_TOTAL.FINAL.ENERGY.CONSUM]"]
gdp_us = df_SE4ALL_GDP_us_years.groupby("Date").sum()["GDP"]

re_share_of_TFEC_us = re_consumption_TJ_us / tfec_TJ_us * 100

re_share_of_TFEC_us

```

```python
#find annual er_consumption_TJ_prc, tfec_TJ_prc, re_share_of_TFEC_prc
#limit df to only China
df_SE4ALL_GDP_prc = df_SE4ALL_GDP[df_SE4ALL_GDP["Country Name"] == "China"]

#limit df to range of years from 1990 to 2015
df_SE4ALL_GDP_prc_years = df_SE4ALL_GDP_prc[df_SE4ALL_GDP_prc["Date"].between(1990, 2015)]

#group by year and grab relevant data
re_consumption_TJ_prc = df_SE4ALL_GDP_prc_years.groupby("Date").sum()["Renewable energy consumption (TJ) [3.1_RE.CONSUMPTION]"]
tfec_TJ_prc = df_SE4ALL_GDP_prc_years.groupby("Date").sum()["Total final energy consumption (TFEC) (TJ) [1.1_TOTAL.FINAL.ENERGY.CONSUM]"]
gdp_prc = df_SE4ALL_GDP_prc_years.groupby("Date").sum()["GDP"]

re_share_of_TFEC_prc = re_consumption_TJ_prc / tfec_TJ_prc * 100

re_share_of_TFEC_prc
```

```python
# Setting the positions and width for the bars
pos = list(range(len(re_share_of_TFEC_prc))) 
width = 0.25

list_of_years = range(1990, 2016)

# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using re_share_of_TFEC_eu data,
        re_share_of_TFEC_eu, 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.7, 
        # with color
        color='#FFCC00', 
        # with label the first value in first_name
        label="EU") 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using re_share_of_TFEC_us data,
        re_share_of_TFEC_us, 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.7, 
        # with color
        color='#3c3b6e', 
        # with label the first value in first_name
        label="US") 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        re_share_of_TFEC_prc, 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.7, 
        # with color
        color='#aa381e', 
        # with label the third value in first_name
        label='China') 

# Set the y axis label
ax.set_ylabel('Percent')

ax.set_xlabel('Year')

# Set the chart's title
ax.set_title('Renewable Energy Use as % of Total Final Energy Consumption Annually')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(list_of_years)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(re_share_of_TFEC_prc)*1.10] )

# Adding the legend and showing the plot
plt.legend(['EU', 'US', 'China'], loc='upper right')

plt.show()
```

```python
plt.plot(list_of_years, re_share_of_TFEC_eu , marker='', color='#FFCC00', linewidth=2)
plt.plot(list_of_years, re_share_of_TFEC_us , marker='', color='#3c3b6e', linewidth=2)
plt.plot(list_of_years, re_share_of_TFEC_prc, marker='', color='#aa381e', linewidth=2)

# Set the y and y axis labels
plt.ylabel('Percent')
plt.xlabel('Year')

# Set the chart's title
plt.title('Renewable Energy Use as % of \nTotal Final Energy Consumption Annually, 1990-2015')
plt.legend(['EU', 'US', 'China'], loc='upper right')
```

```python
plt.plot(list_of_years, re_consumption_TJ_eu/1000 , marker='', color='#FFCC00', linewidth=2)
plt.plot(list_of_years, re_consumption_TJ_us/1000 , marker='', color='#3c3b6e', linewidth=2)
plt.plot(list_of_years, re_consumption_TJ_prc/1000, marker='', color='#aa381e', linewidth=2)

# Set the y and y axis labels
plt.ylabel('Petajoules')
plt.xlabel('Year')

# Set the chart's title
plt.title('Total Renewable Energy Consumption (PJ), 1990-2015')
plt.legend(['EU', 'US', 'China'], loc='lower right')
```
