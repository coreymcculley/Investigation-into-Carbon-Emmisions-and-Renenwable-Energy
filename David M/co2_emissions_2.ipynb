```python
country = []
date = []
CO2 = []

for index, row in remuv_more.iterrows():
    for i in range(3,len(row)):
        country.append(row["Country"])
        date.append(1971+i-3)
        CO2.append(row[1971+i-3])
                   
    
```

```python
df_dict = {
    "Country Name": country, 
    "Date": date, 
    "CO2 emissions (million tons)": CO2
}

long_df = pd.DataFrame(df_dict)
```

```python
file = os.path.join("..","Source Databases","Country Codes.csv")
df_country_code = pd.read_csv(file)
df_country_code
```

```python
df_country_code = df_country_code[["name","alpha-3"]]
```

```python
long_df
```

```python
df_country_code = df_country_code.rename(columns={"alpha-3": "Country Code", "name": "Country Name"})
df_country_code["Lat"] = ""
df_country_code["Lng"] = ""
df_country_code

```

```python
params = {"key": gkey} 

for index, row in df_country_code.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    country = row["Country Name"]
    print(f"Searching within {country}")


    # update address key value
    params['address'] = country

    # Execute request
    country_lat_lng = requests.get(base_url, params=params)
    
    # convert response to json
    country_lat_lng = country_lat_lng.json()

    # Add resulting lat & lng to their respective columns at the current iteration's row index
    df_country_code.loc[index, "Lat"] = country_lat_lng["results"][0]["geometry"]["location"]["lat"]
    df_country_code.loc[index, "Lng"] = country_lat_lng["results"][0]["geometry"]["location"]["lng"]
```

```python
df_country_code
```

```python
merge_table = pd.merge(long_df, df_country_code, on="Country Name", how="left")
merge_table

```

```python
merge_table_drop = merge_table.dropna()
```

```python
merge_table_drop
```

```python
merge_table_drop_2015 = merge_table_drop[merge_table_drop["Date"] == 2015]
```

```python
problem=merge_table_drop_2015[merge_table_drop_2015["CO2 emissions (million tons)"]==".."].index
merge_table_drop_2015=merge_table_drop_2015.drop(problem)
merge_table_drop_2015["CO2 emissions (million tons)"] = merge_table_drop_2015["CO2 emissions (million tons)"].astype("float")
```

```python
locations = merge_table_drop_2015[["Lat","Lng"]]
weights = merge_table_drop_2015["CO2 emissions (million tons)"].astype(float)

# create the mapping figure
fig = gmaps.figure()

# create the heatmap layer
heat_layer = gmaps.heatmap_layer(locations, weights=weights)
heat_layer.max_intensity = 100
heat_layer.point_radius = 20

# add the heatmap layer to the figure
fig.add_layer(heat_layer)

# display the figure
fig
```
