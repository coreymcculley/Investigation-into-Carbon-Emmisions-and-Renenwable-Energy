```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import os
```

```python
co2_data_path = "clean_co2_emissions_data.csv"
co2_data = pd.read_csv(co2_data_path)
co2_data
```

```python
middle_east_df = co2_data.loc[co2_data["Region/Country/Economy"] == "Middle East"]
middle_east_df.reset_index(drop=True).head()
```

```python
x_axis = middle_east_df["Year"]
middle_east_mtco2e = middle_east_df["Million tonnes of CO2"].astype(float)

plt.figure(figsize=(15,10))

plt.plot(x_axis,middle_east_mtco2e,linewidth=2,marker="",color="red")

plt.title("Total CO2 emissions contributed by the Middle East from 1971 to 2017")
plt.xlabel("Year")
plt.ylabel("Million tonnes of CO2")

output_path = os.path.join("Charts", "middleeastregionstotalco2emissions.png")
plt.savefig(output_path)
plt.show()
```

```python
years = range(1970, 2017)

plt.figure(figsize=(12,12))
plt.plot(years, middle_east_mtco2e,  marker='', color='red', linewidth=3,label="Middle East")
plt.plot(years, non_oecd_americas_mtco2e, marker='', color='orange', linewidth=3,label="Non-OECD Americas")
plt.plot(years, china_only_mtco2e, marker='', color='yellow', linewidth=3,label="China")
plt.plot(years, not_china_mtco2e, marker='', color='green', linewidth=3,label="Asia (does not include China)")
plt.plot(years, africa_mtco2e, marker='', color='blue', linewidth=3,label="Africa")
plt.plot(years, non_europe_eurasia_mtco2e, marker='', color='indigo', linewidth=3,label="Non-OECD Europe and Eurasia")
plt.plot(years, europe_mtco2e, marker='', color='violet', linewidth=3,label="OECD Europe")
plt.plot(years, oceania_mtco2e, marker='', color='black', linewidth=3,label="OECD Asia Oceania")
plt.plot(years, americas_mtco2e, marker='', color='grey', linewidth=3,label="OECD Americas")
plt.legend(frameon=True, labelspacing=1, title='Regions')
plt.title("Total CO2 emissions contributed by all countries (OECD and non OECD) from 1971 to 2017")
plt.xlabel("Year")
plt.ylabel("Million tonnes of CO2")

output_path = os.path.join("Charts", "allregionstotalco2emissions.png")
plt.savefig(output_path)
plt.show()
```
