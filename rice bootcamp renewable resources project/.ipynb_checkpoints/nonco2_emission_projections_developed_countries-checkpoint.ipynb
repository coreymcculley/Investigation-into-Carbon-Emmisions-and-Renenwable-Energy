```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import os
```

```python
nonco2_data_path = 'clean_non_co2_emissions_data.csv'
nonco2_data = pd.read_csv(nonco2_data_path)
nonco2_data
```

```python
list_of_developed_countries_eu = ["Austria", "Belgium", "Denmark", "Finland", "France", "Germany", "Greece", "Ireland", "Italy", "Luxembourg", "Netherlands", "Portugal", "Spain", "Sweden", "United Kingdom", 
                                  "Bulgaria", "Croatia", "Cyprus", "Czech Republic", "Estonia", "Hungary", "Latvia", "Lithuania", "Malta", "Poland", "Romania", "Slovak Republic", "Slovenia", 
                                  "Iceland", "Norway", "Switzerland"]
list_of_developed_countries_other = ["Australia", "Canada", "Japan", "New Zealand", "United States"]
list_of_developed_countries = list_of_developed_countries_eu + list_of_developed_countries_other
```

```python
developed_countries = nonco2_data[nonco2_data["Country"].isin(list_of_developed_countries)]
developed_countries.reset_index(drop=True)
```

```python
projectionsfor1990 = developed_countries.loc[developed_countries["Year"] == 1990]
projectionsfor1990.plot.barh(x="Country", y="MtCO2e",figsize=(15,10))

plt.title("Total Non-CO2 emissions contributed by developed countries in 1990",fontsize=18)
plt.xlabel("Metric tons of carbon dioxide equivalent",fontsize=18)
plt.ylabel("Country",fontsize=18)

output_path = os.path.join("Charts", "developedcountries1990projections.png")
plt.savefig(output_path)
plt.show()
```
