```python
remuv_ending_rows = df.drop(df.index[190:194])
remuv_ending_rows.reset_index(drop=True)
```

```python
remuv_more = remuv_ending_rows.drop(remuv_ending_rows.index[0:23])
remuv_more.reset_index(drop=True)
```

```python
rename_1st_column = remuv_more.rename(columns={"million tonnes of CO2": "Region/Country/Economy"})
rename_1st_column
```

```python
final_co2_data = rename_1st_column.melt(id_vars=["Region/Country/Economy"],var_name="Year",value_name="Million tonnes of CO2")
final_co2_data
```

```python
final_co2_data.to_csv('co2_emissions.csv')
```

```python
co2_data_path = "co2_emissions.csv"
```

```python
co2_data = pd.read_csv(co2_data_path)
```

```python
co2_data
```

```python
#regions = ["Middle East","Non-OECD Americas","China (incl. Hong Kong, China)","Asia (excl. China)","Africa","Non-OECD Europe and Eurasia","OECD Europe","OECD Asia Oceania","OECD Americas"]
```

```python
middle_east_df = co2_data.loc[co2_data["Region/Country/Economy"] == "Middle East"]
middle_east_df.reset_index(drop=True).head()
```

```python
x_axis = middle_east_df["Year"]
mtco2e = middle_east_df["Million tonnes of CO2"]

plt.figure(figsize=(15,10))

plt.plot(x_axis,mtco2e,linewidth=1,marker="o",color="red")

plt.title("Total CO2 emissions contributed by the Middle East from 1971 to 2017")
plt.xlabel("Year")
plt.ylabel("Million tonnes of CO2")

plt.show()
```
