```python
import pandas as pd
import requests
import json
from pprint import pprint
import os
```

```python
url="http://api.worldbank.org/v2/country/USA/indicator/NY.GDP.MKTP.PP.CD?format=json&date=1980:2019"
#more indicators can be found here https://datacatalog.worldbank.org/harvest-source/indicators-api
#the rest is straight forward, more info can be found here https://datahelpdesk.worldbank.org/knowledgebase/articles/898581-api-basic-call-structures
results=requests.get(url).json()

pprint(results)
```

```python
WB_Data=[]

for entry in results[1]:
    yearDict={"Country Code":entry["countryiso3code"],"Date":entry["date"],"GDP":entry["value"]}
    WB_Data.append(yearDict)
WB_DF=pd.DataFrame(WB_Data)
WB_DF
```

```python
WB_DF.dropna(inplace=True)
WB_DF
```

```python
countryList=["DEU","FRA","ITA","SWE","POL","ESP","NLD","GRC","ROU","HRV","DNK","IRL","BGR","BEL","PRT","AUT","CZE","FIN","HUN","LUX",
             "CYP","LTU","SVN","MLT","EST","LVA","SVK","USA","GBR","CHN"]
```

```python
#Gets GDP info and puts it in a data frame called WB_DF
WB_Data=[]
for country in countryList:
    url=f"http://api.worldbank.org/v2/country/{country}/indicator/NY.GDP.MKTP.PP.CD?format=json&date=1990:2019"

    results=requests.get(url).json()
    for entry in results[1]:
        yearDict={"Country Code":entry["countryiso3code"],"Date":entry["date"],"GDP":entry["value"]}
        WB_Data.append(yearDict)
WB_DF=pd.DataFrame(WB_Data)
```

```python
#Gets GDP info and puts it in a data frame called PerCapDF
WB_Data=[]
for country in countryList:
    url=f"http://api.worldbank.org/v2/country/{country}/indicator/NY.GDP.PCAP.PP.KD?format=json&date=1990:2019"

    results=requests.get(url).json()
    for entry in results[1]:
        yearDict={"Country Code":entry["countryiso3code"],"Date":entry["date"],"GDP Per Capita":entry["value"]}
        WB_Data.append(yearDict)
PerCapDF=pd.DataFrame(WB_Data)
```

```python
#Merges WB_DF and PerCapDF
WB_DF=WB_DF.merge(PerCapDF,on=["Country Code","Date"])
```

```python
#Prints out WB_DF as a csv in the same folder
file = os.path.join("WorldBankGDP.csv")
WB_DF.to_csv(file, index=False, header=True)
```
